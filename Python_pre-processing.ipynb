# Algorithm

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans

# ✅ Step 1: Clean & Encode Data
df_encoded = df.copy()

# Map AgeGroup if not numeric
if df_encoded['AgeGroup'].dtype == 'object':
    age_group_map = {'18-25': 0, '26-35': 1, '36-45': 2, '46-55': 3, '56+': 4}
    df_encoded['AgeGroup'] = df_encoded['AgeGroup'].map(age_group_map)

# Label Encoding
label_encoders = {}
for col in df_encoded.select_dtypes(include='object').columns:
    if col != 'Attrition':
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
        label_encoders[col] = le

# Encode Attrition
if df_encoded['Attrition'].dtype == 'object':
    df_encoded['Attrition'] = LabelEncoder().fit_transform(df_encoded['Attrition'])

# Fill missing values
df_encoded = df_encoded.fillna(df_encoded.mean(numeric_only=True))

# ----------------------------
# 🚀 Random Forest Classifier
# ----------------------------
X = df_encoded.drop(['Attrition', 'EmpID'], axis=1)
y = df_encoded['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

print("🔹 Random Forest Results")
print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))

# Feature Importance
importances = pd.Series(rf_model.feature_importances_, index=X.columns)
importances.sort_values(ascending=True).plot(kind='barh', figsize=(10, 8), title="Random Forest Feature Importance")
plt.tight_layout()
plt.show()

# ----------------------------
# 🔍 K-Means Clustering
# ----------------------------
X_kmeans = df_encoded.drop(['EmpID', 'Attrition'], axis=1)

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_kmeans)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded['Cluster'] = kmeans.fit_predict(X_scaled)

# Plotting clusters using PCA for 2D visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)

plt.figure(figsize=(10,6))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=df_encoded['Cluster'], palette='Set2')
plt.title('K-Means Clustering Visualization (PCA Reduced)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```

You can analyze the cluster profiles:

Cluster 0 (Green) might represent mid-level, stable employees.

Cluster 1 (Orange) could be high-risk or high-performing employees.

Cluster 2 (Blue) might represent junior/new employees.

Job Level vs Income:

Monthly income increases proportionally with job level, showing a clear compensation structure. However, employees in lower job levels (especially levels 1 and 2) might feel underpaid, which could contribute to higher attrition.

Attrition & Overtime:
